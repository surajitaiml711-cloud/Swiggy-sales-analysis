# **CHART DESIGN**

# **Monthly Sales Trend**

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.to_period('M').astype(str)
monthly_sales = df.groupby('Month')['Price (INR)'].sum().reset_index()

# fig = plt.figure(figsize=(10, 6))
# plt.plot(monthly_sales['Month'], monthly_sales['Price (INR)'], marker='o', linestyle='-')
# plt.xlabel('Month')
# plt.xticks(rotation=45)
# plt.ylabel('Total Sales (INR)')
# plt.title('Monthly Sales Trend')

fig = px.line(monthly_sales, x='Month', y='Price (INR)', title='Monthly Sales Trend')
fig.update_xaxes(tickangle=45)
fig.show()
```

# **Daily Sales Trend**

```python
df["DayName"] = df["Order Date"].dt.day_name()
daily_sales = df.groupby("DayName")["Price (INR)"].sum().reindex(["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])

# fig = plt.figure(figsize=(10, 6))
# plt.plot(daily_sales.index, daily_sales.values)
# plt.xlabel('Day of the Week')
# plt.xticks(rotation=45)
# plt.ylabel('Total Sales (INR)')
# plt.title('Daily Sales Trend')
# plt.show()

fig = px.bar(daily_sales, x=daily_sales.index, y=daily_sales.values, title="Daily Sales Trend")
fig.update_xaxes(tickangle=45)
fig.show()
```

# **Total Sales by Food Type (Veg vs Non-Veg)**

```python
non_veg = [
    "chicken","egg","fish","kabab","kebab","mutton","biryani","Non-veg","Non veg","prawn","biriyani"
]

df["Food Category"] = np.where(df["Dish Name"].str.lower().str.contains('|'.join(non_veg),na=False), "Non-Veg", "Veg")
```

```python
food_revenue = df.groupby("Food Category")["Price (INR)"].sum().reset_index()
```

```python
import plotly.express as px

fig = px.pie(
    food_revenue,
    values="Price (INR)",
    names="Food Category",
    title="🍽️ Total Sales by Food Type",
    hole=0.45,   # Donut chart
    color="Food Category",
    color_discrete_map={
        "Veg": "#2ca02c",
        "Non-Veg": "#d62728"
    }
)

fig.update_traces(
    textposition="inside",
    textinfo="percent+label+value",
    pull=[0, 0.08],          # Pull out second slice
    marker=dict(line=dict(color="white", width=2)),
    hovertemplate="<b>%{label}</b><br>Sales: ₹%{value:,.0f}<br>Share: %{percent}<extra></extra>"
)

fig.update_layout(
    title={
        "text": "🍽️ Total Sales by Food Type",
        "x": 0.5,
        "xanchor": "center"
    },
    font=dict(size=16),
    legend_title="Food Category",
    template="plotly_white",
    width=700,
    height=500
)

fig.show()
```

# **Total Sales by State (Map Visualization)**

```python
Total_sales_by_state = df.groupby("State",as_index=False)["Price (INR)"].sum().sort_values(by="Price (INR)",ascending=False)
```

```python
import plotly.express as px

fig = px.bar(
    Total_sales_by_state,
    x="Price (INR)",
    y="State",
    color="State",
    orientation="h",
    title="📍 Total Sales by State",
    labels={
        "State": "State",
        "Price (INR)": "Total Sales (INR)"
    }
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    showlegend=False,
    height=600
)

fig.update_traces(
    texttemplate="₹%{x:,.0f}",
    textposition="outside"
)

fig.show()
```

# **Quarterly Performance Summary**

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Quarter'] = df['Order Date'].dt.to_period('Q').astype(str)
quater_summary = df.groupby('Quarter',as_index=False).agg(
    Total_sales = ("Price (INR)","sum"),
    Average_rating = ("Rating","mean"),
    Total_orders = ("Order Date","count")
).sort_values(by="Quarter")

quater_summary["Total_sales"] = quater_summary["Total_sales"].apply(lambda x : round(x,0))
quater_summary["Average_rating"] = quater_summary["Average_rating"].apply(lambda x : round(x,2))


quater_summary
```

# **Weekly Trend Analysis**

```python
weekly_trend = (
    df.groupby(df["Order Date"].dt.isocalendar().week)["Price (INR)"]
      .sum()

)

```

```python
fig = px.line(
    weekly_trend,
    x=weekly_trend.index,
    y=weekly_trend.values,
    title="Weekly Trend Analysis",
    labels={"x": "Week", "y": "Total Sales (INR)"}
)

fig.update_layout(
    xaxis_title="Week",
    yaxis_title="Total Sales (INR)",
    template="plotly_white"
)

fig.show()
```
